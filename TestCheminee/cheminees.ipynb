```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python

import numpy as np
fig,ax = plt.subplots()
data = np.genfromtxt('Cheminees.csv', delimiter=';', names=True)
L = data["L"]
R = data["ReR"] + 1j*data["ImR"]
T = data["ReT"] + 1j*data["ImT"]
RN = R + T
RD = R - T
ax.plot(np.cos(np.linspace(0,2*np.pi,100)),np.sin(np.linspace(0,2*np.pi,100)),color="lightgrey")
ax.plot(np.real(RN),np.imag(RN),label="RN",linestyle="",marker=".")
ax.plot(np.real(R),np.imag(R),label="R",linestyle="",marker=".")
ax.plot(np.real(T),np.imag(T),label="T",linestyle="",marker=".")
ax.set_aspect(1)
ax.legend()
plt.show()
```

```python
plt.plot(L,-np.log(np.abs(T-1)))
plt.show()
```
