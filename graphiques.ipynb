```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

realList = []
imagList = []
with open("R.txt","r") as f:
    for line in f.readlines():
        tup = tuple(line[1:-2].split(sep=","))
        realList.append(float(tup[0]))
        imagList.append(float(tup[1]))
Real = np.array(realList)
Imag = np.array(imagList)
Module = np.sqrt(Real**2 + Imag**2)

plt.plot(Module, label = 'Module')
plt.plot(Real, label = "Partie réelle")
plt.plot(Imag,label="Partie imaginaire")
plt.legend()
plt.plot()
```
