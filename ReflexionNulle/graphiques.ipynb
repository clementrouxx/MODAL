```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

realList = []
imagList = []
with open("R.txt","r") as f:
    for line in f.readlines():
        tup = tuple(line[1:-2].split(sep=","))
        realList.append(float(tup[0]))
        imagList.append(float(tup[1]))
Real = np.array(realList)
Imag = np.array(imagList)
Module = np.sqrt(Real**2 + Imag**2)

plt.axhline(0,color="black")

plt.plot(Module, label = '|R|')
plt.plot(Real, label = "Re(R)")
plt.plot(Imag,label="Im(R)")
plt.xlabel("It√©rations")
plt.legend()
plt.plot()
```
